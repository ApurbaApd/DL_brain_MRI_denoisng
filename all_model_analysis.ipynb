```python
noise_levels=[0.04,0.06,0.08,0.10]

def plot_metric(metric):
    plt.figure(figsize=(10,6), dpi=300)
    metric_idx=["PSNR","SSIM","EPI"].index(metric)
    for model in models:
        for noise in noise_levels:
            folds=np.array(data[noise][model])

            for i in range(len(folds)):
                plt.scatter(
                    noise,
                    folds[i,metric_idx],
                    alpha=0.25
                )

        sub=df[df.Model==model].sort_values("Noise")

        plt.plot(
            sub["Noise"],
            sub[f"{metric}_mean"],
            marker='o',
            linewidth=3,
            label=model
        )

    plt.xlabel("Noise Factor ($\sigma$)")
    plt.ylabel(metric)
    plt.xticks(noise_levels)
    
    plt.tick_params(axis='both', labelsize=30)
    plt.legend()
    plt.tight_layout()
    plt.grid(False)
 
    plt.show()


plot_metric("PSNR")
plot_metric("SSIM")
plot_metric("EPI")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Mean matrix (used for colors)
mean_pivot = df.pivot(index="Model", columns="Noise", values="PSNR_mean")

# Std matrix
std_pivot = df.pivot(index="Model", columns="Noise", values="PSNR_std")

# Build annotation labels
labels = mean_pivot.copy().astype(str)

for i in range(mean_pivot.shape[0]):
    for j in range(mean_pivot.shape[1]):
        mean_val = mean_pivot.iloc[i, j]
        std_val = std_pivot.iloc[i, j]

        labels.iloc[i, j] = f"{mean_val:.2f}\n±{std_val:.2f}"

plt.figure(figsize=(9,5), dpi=300)

sns.heatmap(
    mean_pivot,
    annot=labels,
    #increase size
    fmt="",
    cmap="viridis",
    cbar_kws={'label': 'PSNR (dB)'},
    linewidths=0.5,
    annot_kws={"size": 22}
)

plt.title("PSNR (Mean ± Std) across folds")
plt.xlabel("Noise Factor ($\sigma$)")
plt.ylabel("Models")

plt.tick_params(axis='x', labelsize=27)
plt.tick_params(axis='y', labelsize=18)

plt.tight_layout()
plt.savefig(f'')
plt.show()
```
